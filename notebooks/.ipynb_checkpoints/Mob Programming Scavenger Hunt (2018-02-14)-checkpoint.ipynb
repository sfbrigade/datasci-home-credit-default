```python
import os
import sys

import pandas as pd
pd.options.display.max_columns = 999
import sqlalchemy as sqla
from sqlalchemy import create_engine

DB_URI = os.getenv('CD_DWH')
engine = create_engine(DB_URI)
```

Below are the tables that currently exist in the database!

```python
QUERY = """
select *
from information_schema.tables
where table_schema not in ('information_schema', 'pg_catalog', 'public')
"""
with engine.begin() as conn:
    results = pd.read_sql(QUERY, conn)
results.head(100)
```

## Query Example

```python
QUERY = """
select
  *
from trg_analytics.candidate_contributions
"""
with engine.begin() as conn:
    results = pd.read_sql(QUERY, conn)
results.head(100)
```

```python
results.head()
```

```python
# Question 1: Number of rows in a Pandas DataFrame
len(results.index)
```

```python
results.index
```

```python
# Question 2: What is the total sum of donations made to candidates in the 2017-2018 election cycle?
election_cycle_2017 = results.loc[results.election_cycle=='2017']
```

```python
election_cycle_2017.transaction_amount.sum()
```

```python
# Question 2b: Bar Graph of sum of donations by year
donations_by_year = results.groupby(['election_cycle'])['transaction_amount'].sum()
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
plt.bar(donations_by_year.index, donations_by_year)
```

```python
# Across history what are the top 5 zip codes ranked according to the donations sourced from those areas?
# Which cities are these zip codes located in? Do you notice anything surprising?
results.groupby(['donor_zip_code'])['transaction_amount'].sum().sort_values(ascending=False).head()
```

```python
# In the 2015-2016 election cycle, which candidate received the most donations?
election_cycle_2015 = results.loc[results.election_cycle=='2015']
```

```python
results.groupby(['recipient_candidate_name'])['transaction_amount'].sum().sort_values(ascending=False).head()
```

```python
# In the 2017-2018 election cycle, how many donors comprised of 50% of all donations made (by dollar amount)?
donors_2017 = election_cycle_2017.groupby(['donor_name'])['transaction_amount'].sum().to_frame()
```

```python
donors_2017.describe()
```

```python
donors_2017 = donors_2017.sort_values(by='transaction_amount')
```

```python
donors_2017['cum_sum'] = donors_2017.transaction_amount.cumsum()
```

```python
donors_2017['cum_perc'] = 100*donors_2017.cum_sum/donors_2017.transaction_amount.sum()
```

```python
donors_2017.loc[donors_2017.cum_perc >= 50].shape
```
